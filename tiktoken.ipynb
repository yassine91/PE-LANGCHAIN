```python
import tiktoken
```

```python
# create a tokenizer for gpt-4o
tokenizer = tiktoken.encoding_for_model("gpt-4o")
# could be also used : 
# tokenizer = tiktoken.get_encoding("o200_base")

# print the name of tokenizer
print(tokenizer.name)
```

```python
prompt = "Vous êtes un assistant expert dans le domaine des analyses de sentiments"
tokens =tokenizer.encode(prompt)
print(len(tokens))

for token in tokens:
    t = tokenizer.decode_single_token_bytes(token).decode()
    print(t,end = "|")
```

```python
## Function to calculate the number of tokens in a prompt
def token_count(prompt:str, model:str = "o200k_base"):
    tokenizer = tiktoken.get_encoding(model)
    return len(tokenizer.encode(prompt))
```

```python
## Call of method

token_number  = token_count("Hello how are you ? ")
print(token_number)
```
